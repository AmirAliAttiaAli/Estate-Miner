```python
df["available_from"] = pd.to_datetime(df["available_from"])
```

```python
df.info()
```

```python
df["available_year"] = df["available_from"].dt.year
df["available_month"] = df["available_from"].dt.month
df["available_day"] = df["available_from"].dt.day
```

```python
df[["available_from", "available_year", "available_month", "available_day"]].head()
```

```python
categorical_cols = [
    "property_type",
    "region",
    "sub_area",
    "furnishing_status"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df.info()
```

```python
df["price_per_sqm"] = df["price"] / df["area_sqm"]
```
