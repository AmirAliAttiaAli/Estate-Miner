# Estate Miner Preprocessing

## Notebook Flow

- Load the cleaned dataset and inspect its structure.
- Convert and standardize key columns for analysis.
- Create derived features that summarize price, space, and amenities.

```python
import numpy as np
import pandas as pd
```

## 1. Load and Inspect the Dataset

Start by loading the cleaned listings data and reviewing its basic structure, sample rows, and column layout.

```python
df = pd.read_csv("Data/cleaned_data.csv")
```

```python
df.head()
```

```python
df.shape
```

```python
df.columns
```

```python
df.info()
```

```python
df["available_from"] = pd.to_datetime(df["available_from"])
```

```python
df.info()
```

## 2. Create Date-Based Features

Convert the listing date into a datetime field and extract useful calendar components for analysis.

```python
df["available_year"] = df["available_from"].dt.year
df["available_month"] = df["available_from"].dt.month
df["available_day"] = df["available_from"].dt.day
```

```python
df[["available_from", "available_year", "available_month", "available_day"]].head()
```

```python
categorical_cols = [
    "property_type",
    "region",
    "sub_area",
    "furnishing_status"
]

for col in categorical_cols:
    df[col] = df[col].astype("category")
```

```python
df.info()
```

## 3. Standardize Categorical Columns

Cast the main categorical fields to the category dtype to reduce memory usage and make the data easier to work with.

```python
df["price_per_sqm"] = df["price"] / df["area_sqm"]
```

```python
df[["price", "area_sqm", "price_per_sqm"]].head()
```

## 4. Engineer Pricing and Size Features

Create ratio-based features that normalize price and area by bedrooms, bathrooms, and floor area.

```python
amenity_cols = [col for col in df.columns if col.startswith("amenity_")]

df["amenities_count"] = df[amenity_cols].sum(axis=1)
```

```python
df[["amenities_count"]].head()
```

```python
df["amenities_count"].value_counts().sort_index()
```

```python
df[amenity_cols].sum().sort_values(ascending=False)
```

## 5. Build Amenity and Room Features

Summarize available amenities and derive room-based indicators that help compare listings at a glance.

```python
df["total_rooms"] = df["bedrooms"] + df["bathrooms"]
```

```python
df[["bedrooms", "bathrooms", "total_rooms"]].head()
```

```python
(df["bedrooms"] == 0).sum()
```

```python
df["price_per_bedroom"] = df["price"] / df["bedrooms"]
```

```python
df[["price", "bedrooms", "price_per_bedroom"]].head()
```

```python
(df["bathrooms"] == 0).sum()
```

```python
df["price_per_bathroom"] = df["price"] / df["bathrooms"]
```

```python
df[["price", "bathrooms", "price_per_bathroom"]].head()
```

```python
df["area_per_bedroom"] = df["area_sqm"] / df["bedrooms"]
```

```python
df[["area_sqm", "bedrooms", "area_per_bedroom"]].head()
```

```python
df["area_per_bathroom"] = df["area_sqm"] / df["bathrooms"]
```

```python
df[["area_sqm", "bathrooms", "area_per_bathroom"]].head()
```

```python
df["amenities_per_room"] = df["amenities_count"] / df["total_rooms"]
```

```python
df[["amenities_count", "total_rooms", "amenities_per_room"]].head()
```

```python
new_features = [
    "price_per_sqm",
    "amenities_count",
    "total_rooms",
    "price_per_bedroom",
    "price_per_bathroom",
    "area_per_bedroom",
    "area_per_bathroom",
    "amenities_per_room"
]

df[new_features].head()
```
